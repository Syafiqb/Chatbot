```python
#melihat data unique di setiap kolom
pd.set_option('display.max_colwidth', None)
features=[]
unique=[]
nunique=[]
miss_sum=[]

for i in df.columns:
    features.append(i)
    unique.append(df.sort_values(i)[i].unique())
    nunique.append(df[i].nunique())
    miss_sum.append(df[i].isna().sum())

pd.DataFrame({'features':features,
            'unique':unique,
             'nunique':nunique,
             'miss_sum':miss_sum})
```
